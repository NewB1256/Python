```python
""" 11. Write a program to check if a given list contains any duplicates using 
a `for` loop."""

lis = [1,2,5,4,3,4,5,4,3,2]
dup_lis = []
res = False
for i in lis:
    if i in dup_lis:
        res = True
        break  
    else:
        dup_lis.append(i)
        
if res:
    print("Yes! the list contains duplicates")
else:
    print("No! the list doesn't contains duplicates")
```

```python
""" 14. Write a program to find the maximum element in a 2D list using 
a nested `for` loop."""

my_list = [1, [2, 3, 4], 8, [5, 34, 56]]

max_element = None
for element in my_list:
    if type(element) == list:
        for sub_element in element:
            if  max_element is None or (type(sub_element) in (int, float) and sub_element > max_element):
                max_element = sub_element
    elif type(element) in (int, float):
        if max_element is None or element > max_element:
            max_element = element

if max_element is not None:
    print("Maximum element in the nested list is:", max_element)
else:
    print("The list does not contain any numeric elements.")
```

```python
""" 20. Write a program to find the greatest common divisor (GCD) 
of two numbers using a `for` loop."""

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
min_num = min(num1, num2)
gcd = 1

for i in range(1, min_num + 1):
    if num1 % i == 0 and num2 % i == 0:
        gcd = i

print(f"The GCD of {num1} and {num2} is:", gcd)
```

```python
""" 31. Write a program to generate a list of prime numbers using 
the Sieve of Eratosthenes algorithm and list comprehension."""

n = int(input("Enter the upper limit (n) for prime numbers: "))
is_prime = [True] * (n + 1)
is_prime[0] = is_prime[1] = False
for p in range(2, int(n**0.5) + 1):
    if is_prime[p]:
        for i in range(p * p, n + 1, p):
            is_prime[i] = False

prime_numbers = [i for i in range(2, n + 1) if is_prime[i]]

print("Prime numbers up to", n, "are:", prime_numbers)
```

```python
""" 32. Create a program that generates a list of all Pythagorean 
triplets up to a specified limit using list comprehension."""

n = int(input("Enter the upper limit for Pythagorean triplets: "))
triplets = [(a, b, c) for a in range(1, n + 1) for b in range(a, n + 1) for c in range(b, n + 1) if a**2 + b**2 == c**2]

print("Pythagorean triplets up to", n, "are: ",triplets)
```

```python
""" 34. Write a program that calculates the mean, median, and mode 
of a list of numbers using list comprehension."""

num = [float(x) for x in input("Enter a list of numbers (comma-separated values): ").split(',')]
a = sum(num)
b = len(num)
mean = a / b
median = sorted(num)[len(num) // 2]
mode = max(num, key = num.count)

# Print the results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
""" 38. Create a program that filters a list of strings to include only 
those with more than three vowels using list comprehension."""

strings = input("Enter a list of strings (comma-separated values): ").lower().split(',')

filtered_strings = [s for s in strings if len([char for char in s if char in "aeiou"]) > 3]

print("Strings with more than three vowels:")
for s in filtered_strings:
    print(s.strip()) 
```
